4. Investigate how Laplace Smoothing affects training with small sample sizes
5. Test if two categorical variables are independent of each other
6. Infer conditional probability distributions 
7. Test if two samples have the same conditional probability distribution

## Single Categorical Variable Inference

For analysis it is more convenient to use $Z_X$ the **one-hot encoded** version of $X$

```python
Z_x=pd.get_dummies(X)
print("Z_x.shape =",Z_x.shape)
D=Z_x.shape[1]
Z_x.head()
```

So far we were working with a pandas **DataFrame** object. It has a lot of convenient functions, to work on data by column, and can do basic analytics of
the data.

To do more sophisticated analysis we will want to work with the underlying numpy **array**. It is just a $N\times D$ array of numbers, but supports many more flexible ways to do computation with the data at a low level

```python
Z_x=Z_x.as_matrix()
print("Z_x.shape =",Z_x.shape)
Z_x[:5]
```

### Maximum Likelyhood Frequency Estimation

The maximum likelihood estimate of the probability of category $d$ 
is the sample mean of column $d$ in $Z$ array
$$
    p_X(x=d)= \frac{1}{N}\sum_i Z_{i,d}
$$

```python
p_X=Z_x.mean(axis=0)
for i in range(len(X_labels)):
    print(X_labels[i],p_X[i])
```

### Smoothed Bayes Estimation

The Laplace smoothed estimator is
$$
    p_{\textrm{sm}}(X=d)= \frac{\hat{N}_d+1}{N+D}
$$
where the empirical counts $\hat{N}_d$ are
$$
    \hat{N}_d = \sum_i Z_{i,d}
$$

Generate empirical counts

```python
N_hat=Z_x.sum(axis=0)
N_hat
```

Add Laplace Smooting

```python
N_hat=N_hat+1
N_hat
```

Compute Smoothed Frequencies

```python
p_X_smoothed=N_hat/N_hat.sum()
for i in range(len(X_labels)):
    print(X_labels[i],p_X_smoothed[i])
```

Smoothed and Maximum Likelihood probabilities are very close to each other because $N=1000$ is large compared to the number of classes. 

As a rule of thumb, if the empirical count $\hat{N}_d$ are larger than 5, the impact of smoothing is minimal.

### Test X Distribution

Let's read some test sample data to check if they have the  same distribution as training data

```python
Z_x_test1=pd.get_dummies(X_test1)
Z_x_test1.head()
```

We generated the one-hot encodings as before

```python
Z_x_test1=Z_x_test1.as_matrix()
Z_x_test1[:5]
```

```python
Z_x_test2=pd.get_dummies(X_test2).as_matrix()
Z_x_test3=pd.get_dummies(X_test3).as_matrix()
Z_x_test4=pd.get_dummies(X_test4).as_matrix()
```

Empiricial counts are different, but are they significantly different?

```python
print("test1",Z_x_test1.sum(axis=0))
print("test2",Z_x_test2.sum(axis=0))
print("test3",Z_x_test3.sum(axis=0))
print("test4",Z_x_test4.sum(axis=0))
```

We will use the $\chi^2$ test to determine if the test data was generated from the sample distribution $p_X$ as the training data.

$$
    C^2 = \sum_{d=1}^N \frac{(\hat{O}_{d} - E_{d})^2}{E_d}
$$
is distributed as a $\chi^2_{D-1}$ random variable if
$$
    \hat{O}_d = \sum_i Z_{i,d}
$$
for the $Z_{i,d}$ one-hot encoding of the test data and
$$
    E_d = N p_X(X=d)
$$
where $p_X$ is the probability distribution we determined from the training data and $N$ is the size of the test data.

```python
def C2(Z,p):
    D=len(p)
    N=Z.sum() # number of elements
    obs=Z.sum(axis=0)
    expect=N*p
    df=obs-expect
    df2=df*df
    # make sure we don't get an NAN if expected=0
    c2 = (df2/np.maximum(1e-9,expect)).sum()
    return c2,special.chdtrc(D-1,c2)
```

If we run the test on **training** data, $C^2$ score is zero because observed and expected match exactly.

```python
print("Test","C2","p-value",sep="\t")
print("training",C2(Z_x,p_X))
print("test1",C2(Z_x_test1,p_X))
print("test2",C2(Z_x_test2,p_X))
print("test3",C2(Z_x_test3,p_X))
print("test4",C2(Z_x_test4,p_X))
```

It seems `test1` and `test3` are closer to the same distribution as the training set, but `test2` and `test4` do not

### Test X Distribution with Limited Data

We will repeat the test process now, but training on a much smaller $N_{\small}$ number of datapoints

```python
N_small=60
```

#### Maximum Likelihood Probabilities

Let's compute as before the Maximum Likelihood estimate (**MLE**) of the probabilities,
but using only $N_{\textrm{small}}$ data samples

```python
p_X_small=Z_x[:N_small].mean(axis=0)
for i in range(len(X_labels)):
    print(X_labels[i],p_X_small[i])
```

#### Test of MLE estimate

```python
print("training",C2(Z_x[:N_small],p_X_small))
print("test1",C2(Z_x_test1,p_X_small))
print("test2",C2(Z_x_test2,p_X_small))
print("test3",C2(Z_x_test3,p_X_small))
print("test4",C2(Z_x_test4,p_X_small))
```

There  are less extreme values.

#### Laplace smoothed probabilities

We can perfrom  the tests now with the Laplace smoothed estimates we computed before 

```python
print("Test","C2","p-value",sep="\t")
print("training",C2(Z_x[:N_small],p_X_smoothed))
print("test1",C2(Z_x_test1,p_X_smoothed))
print("test2",C2(Z_x_test2,p_X_smoothed))
print("test3",C2(Z_x_test3,p_X_smoothed))
print("test4",C2(Z_x_test4,p_X_smoothed))
```

We can observe that
1. Training data no longer has a perfect $C^2=0$ score.
1. Tests are directionally correct: `test1` and `test3` have the highest $C^2$ scores.
1. The $\chi^2$ probabilities are still too low for all the test.


## Conditional Variable Inference

Now we will analyze how variable $Y$ depends on variable $X$

```python
Z_y=pd.get_dummies(Y)
print("Z_x.shape =",Z_y.shape)
Z_y.head()
```

Turn objects into pure numpy arrays, not panda frames

```python
Z_y=Z_y.as_matrix()
Z_y[:5]
```

Let's encode the test samples of $Y$ too

```python
Z_y_test1=pd.get_dummies(Y_test1).as_matrix()
Z_y_test2=pd.get_dummies(Y_test2).as_matrix()
Z_y_test3=pd.get_dummies(Y_test3).as_matrix()
Z_y_test4=pd.get_dummies(Y_test4).as_matrix()
```

### Test of Independence between $X$ and $Y$

To test independence between to random variables $X$ and $Y$ we can use a variation of the $\chi^2$ test we used before
$$
    C^2 = \sum_{k=1}^{K}\sum_{d=1}^{D} \frac{(\hat{O}_{k,d} - \hat{E}_{k,d})^2}{\hat{E}_{k,d}}
$$
has a $\chi^2$ distribution with $(K-1)\times (D-1)$ degrees of freedom.

$\hat{O}_{d,k}$ is the number of observations with $Y=k$, and $X=d$, which, in terms of the one hot encoded variables can be expressed as

$$
    \hat{O}_{k,d} = \sum_i Z^Y_{i,k}Z^X_{i,d} = (Z^Y)^{\textrm{T}}*Z^X
$$
where the last expression is a matrix multiplication.


The expected number of observation assuming that the two variables are independent is

$$
    \hat{E}_{k,d} = N\, \hat{p}^Y_k\, \hat{p}^X_d
$$
where $\hat{p}^Y$ and $\hat{p}^X$ are the marginal distributions of $X$ and $Y$ estimated from the data.
$$
    p^Y_k= \frac{1}{N}\sum_{i=1}^N Z^Y_k
$$
$$
    p^X_d= \frac{1}{N}\sum_{i=1}^N Z^X_d
$$

```python
def C2_independence(Z_x,Z_y):
    N=len(Z_x)
    D=Z_x.shape[1]
    K=Z_y.shape[1]
    # p_y has index k
    p_y=Z_y.mean(axis=0)
    # p_x has index d
    p_x=Z_x.mean(axis=0)
    # p will be K*D, with indexes k,d
    p=p_y[:,np.newaxis]*p_x[np.newaxis,:]
    # Z_y has indexes i,k and Z_x has indexes i,d
    #Z will be N*K*D, with indexes i,k,d
    Z=Z_y[:,:,np.newaxis]*Z_x[:,np.newaxis,:]
    # sum over i, left with a K*D matrix
    obs=Z.sum(axis=0) # last two expressions are the same as np.dot(Z_y^T,Z_x)
    # expect
    expect=N*p
    df=obs-expect
    df2=df*df
    c2 = (df2/np.maximum(1e-9,expect)).sum()
    return c2,special.chdtrc((K-1)*(D-1),c2)
```

$X$ and $Y$ are clearly dependent of each other.

```python
C2_independence(Z_x,Z_y)
```

But if we compare the X from one test to Y from a different test, they do look independent 

```python
C2_independence(Z_x_test1,Z_y_test2)
```

## Inference of $P(Y|X)$

We know now that $Y$ depends on $X$, let's estimate $P(Y|X)$ using the same techniques as before

Let's define
$$
Z_{i,k,d} = Z^Y_{i,k}\,Z^X_{i,d}
$$

```python
Z=Z_y[:,:,np.newaxis]*Z_x[:,np.newaxis,:]
print("Z.shape =",Z.shape)
Z[:3]
```

And the empirical counts
$$
    \hat{N}_{k,d}= \sum_{i=1}^N Z_{i,k,d}
$$

```python
N_hat=Z.sum(axis=0)
N_hat
```

We have a lot of columns with low counts, so we better do smoothing

```python
N_hat=N_hat+1
N_hat
```

The marginal counts for variable $X$ are
$$ 
    \hat{N}_d = \sum_i Z^X_{i,d} = \sum_k Z_{k,d} 
$$


```python
N_hat_X=N_hat.sum(axis=0)
N_hat_X
```

Our estimate for the probability for $P(Y=k| X=d)$ is
$$
    P(Y=k|X=d)= \frac{\hat{N}_{k,d}+1}{\hat{N}^X_d+K}
$$

```python
Py_cx=N_hat/N_hat_X
Py_cx
```

### Test Conditional Distribution

We can test, conditional on each value of $X$, if the test sample is well represented by our estimated $P(Y|X)$.

```python
Z_test1=Z_y_test1[:,:,np.newaxis]*Z_x_test1[:,np.newaxis,:]
Z_test2=Z_y_test2[:,:,np.newaxis]*Z_x_test2[:,np.newaxis,:]
Z_test3=Z_y_test3[:,:,np.newaxis]*Z_x_test3[:,np.newaxis,:]
Z_test4=Z_y_test4[:,:,np.newaxis]*Z_x_test4[:,np.newaxis,:]
```

```python
print("Label","Test","C2","p-value",sep="\t")
for i1 in range(D):
    p=Py_cx[:,i1]
  
    print(X_labels[i1])
    print("\t","training",C2(Z[:,:,i1],p))
    print("\t","test1",C2(Z_test1[:,:,i1],p))
    print("\t","test2",C2(Z_test2[:,:,i1],p))
    print("\t","test3",C2(Z_test3[:,:,i1],p))
    print("\t","test4",C2(Z_test4[:,:,i1],p))
    print()
```
